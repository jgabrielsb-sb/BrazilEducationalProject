### 2. Teacher Effort ###

```python
### Passing through files of each year and adding it into the list ###
def get_dfs(folder_path, number_columns,skip_rows):
    list_dfs = []
    for file in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file)
        df = pd.read_excel(file_path,skiprows=skip_rows, header=None)

        ### years 2013 2014, 2015, 2016 have a different pattern than the other. Then, treat it in a different manner ###
        if ('2013' in file) or ('2014' in file) or ('2015' in file) or ('2016') in file:
            df = df.drop(columns=1,axis=0)
            df.columns = [i for i in range(number_columns)]

        list_dfs.append(df)

    return list_dfs
```

```python
### will store a unique dataframe with information of all years ###
def get_unique_df(list_dfs):
    unique_df = list_dfs[0]

    ### passing through dataframes and concatenating all of them ###
    for i in range(1, len(list_dfs)):
        unique_df = pd.concat([unique_df, list_dfs[i]], axis=0)

    return unique_df
```

```python
### putting the unique_df into the format we want ###
def get_df_format(unique_df, columns):
    new_df = unique_df.copy()
    
    ### renaming columns ###
    new_df.columns = columns

    ### removing columns (obs: we must rename columns that we want to remove as ...'rem'...)
    columns_to_remove = [column for column in columns if 'rem' in column]
    new_df = new_df.drop(columns=columns_to_remove, axis=0)

    ### removing thash rows ###
    new_df = new_df[new_df["region"].notnull()]
    
    return new_df
```

* define folder_path;
* pass folder_path to get_dfs(folder_path) --> return list_dfs;
* pass list_dfs to get_unique_df() --> return unique_df;
* pass unique_df to get_df_format() --> return new_df;
* save new_df as a csv file.


```python
def get_df(path, skip_rows, columns, number_columns):
    list_dfs = get_dfs(path, number_columns,skip_rows)
    df = get_unique_df(list_dfs)
    df = get_df_format(df, columns)
    return df
```

```python
### mapping values there are different of our defined pattern ###
map_region =  {'RO': 'Rondônia', 'AC': 'Acre','AM': 'Amazonas',
                         'RR': 'Roraima', 'PA': 'Pará','AP': 'Amapá','TO': 'Tocantins',
                         'MA': 'Maranhão','PI': 'Piauí','CE': 'Ceará','RN': 'Rio Grande do Norte',
                         'PB': 'Paraíba','PE': 'Pernambuco', 'AL': 'Alagoas','SE': 'Sergipe',
                         'BA': 'Bahia', 'MG': 'Minas Gerais','ES': 'Espírito Santo',
                         'RJ': 'Rio de Janeiro','SP': 'São Paulo','PR': 'Paraná','SC': 'Santa Catarina',
                         'RS': 'Rio Grande do Sul', 'MS': 'Mato Grosso do Sul','MT': 'Mato Grosso',
                         'GO': 'Goiás', 'DF': 'Distrito Federal'}

map_adm = {"Publico": "Pública"}
```

```python
### function to map values there are possibly diffent from our defined pattern ###
def map_values(df):
    new_df = df.copy()

    new_df['region'] = new_df['region'].replace(map_region)
    new_df['adm'] = new_df['adm'].replace(map_adm)
    return new_df

### check if there are differente values: if yes, return true. if no, return false ###
def check_different_values(df):
    regions_to_sub = [value for value in df['region'].unique() if value not in correct_regions]
    locations_to_sub = [value for value in df['location'].unique() if value not in correct_location]
    adms_to_sub = [value for value in df['adm'].unique() if value not in correct_adm]

    sub = [regions_to_sub, locations_to_sub, adms_to_sub]
    if sum([len(element) for element in sub]) != 0:
        return [regions_to_sub, locations_to_sub, adms_to_sub]
    
    return False
```

```python
### mapping different values ###
studentPerClass_df = map_values(studentPerClass_df)   

### checking if there are diffent values ###
print(check_different_values(studentPerClass_df))  #those different values are header. then, we should remove it

```
