```python
### creating a list with the regions ###
list1 = ['Brasil', 'Norte', 'Sul', 'Centro-Oeste', 'Sudeste', 'Nordeste']
regions = sorted([region for region in adequacy_df['region'].unique() if region not in list1])
regions = list1 + regions

### creating a dict to map region as str to region id ###
map_regions = {region:(i+1) for i, region in enumerate(regions)}
map_regions
```

```python
### creating dict to map location and adm as a int id ###
map_location = {"Total": 1, "Urbana": 2,"Rural": 3 }
map_adm = {'Total': 1, 'Federal': 2, 'Estadual': 3, 'Municipal': 4, 'Privada': 5, 'Pública':6}
```

```python
adequacy_df['region'].unique()
```

```python
### replacing str values as int id values ###
adequacy_df["region"] = adequacy_df["region"].replace(map_regions)
adequacy_df["location"] = adequacy_df["location"].replace(map_location)
adequacy_df["adm"] = adequacy_df["adm"].replace(map_adm)

adequacy_df.head(2)
```

```python
### using melt() to change dataframe format ###
adequacy_df = adequacy_df.melt(id_vars=['year', 'region', 'location', 'adm'],value_name='measurement')
adequacy_df.head(2)
```

```python
### setting our id's to education an

### function to determine education_id (1 for elem ed, 2 for primary ed, 3 for second ed)###
def get_education_id(variable):
    if variable.startswith('elem'):
        return 1
    elif variable.startswith('prim'):
        return 2
    elif variable.startswith('sec'):
        return 3
    else: # 0 is the id for 'Total'
        return 0

### function to determine measurement_group id (1 for G1, 2 for G2, 3 for G3, 4 for G4, 5 for G5) ###
def get_variable_group(variable):
    # Split the string on '_' and take the second part as the variable group number
    parts = variable.split('_')
    if len(parts) > 1:
        return int(parts[1])
    else:
        return 0
    
### creating our id columns (education_id and measurement_group) ###
adequacy_df['education_id'] = adequacy_df['variable'].apply(get_education_id)
adequacy_df['meas_group'] = adequacy_df['variable'].apply(get_variable_group)
adequacy_df['study_id'] = 1
```

```python
adequacy_df = pd.merge(adequacy_df, studyMeas_df, on=['study_id', 'meas_group'])
#adequacy_df = adequacy_df.drop(columns=['meas_name'])
adequacy_df.head(2)
```

### 2. Teacher Effort ###

```python
### getting df by opening csv and removing necessary columns ###
def get_df(file_path):
    df = pd.read_csv(file_path)
    df = df.drop(columns=['Unnamed: 0'])
    return df
```

```python
### mapping 'region', 'location' and 'adm' columns ###
def mapping_df(df, has_location):
    new_df = df.copy()

    new_df["region"] = new_df["region"].replace(map_regions)
    if has_location:
        new_df["location"] = new_df["location"].replace(map_location)
    else:
        new_df["location"] = 1 # location = 'Total'

    
    new_df["adm"] = new_df["adm"].replace(map_adm)

    return new_df
```

```python
studyMeas_df
```

```python
### melting dataframe ###
def melt_df(df, melt_columns):
    new_df = df.copy()
    new_df = new_df.melt(id_vars=melt_columns,value_name='measurement')
    return new_df
```

```python
### function to get education_id ###
def get_education_id2(df):
    new_df = df.copy()
    new_df['education_id'] = new_df['variable'].apply(get_education_id)

    return new_df

### function to get study id ###
def get_study_id(df, id):
    new_df = df.copy()
    new_df['study_id'] = id
    return new_df

### function to get measurement group ###
def get_meas_group(df):
    new_df = df.copy()
    new_df['meas_group'] = new_df['variable'].apply(get_education_id)

    return new_df
```

```python
### get study_meas_id ###
def get_studyMeas_id(df):
    new_df = df.copy()
    new_df = pd.merge(new_df, studyMeas_df, on=['study_id', 'meas_group'])
    return new_df

### removing columns that don't match measurements table ###
def remove_columns(df):
    new_df = df.copy()
    new_df = new_df.drop(columns=columns_to_remove,axis=0)
    return new_df

### renaming columns ###
def rename_columns(df):
    new_df = df.copy()
    new_df = new_df[['year', 'region', 'location', 'adm', 'measurement', 'education_id','study_meas_id']]
    new_df.columns = ['year', 'region_id', 'location_id', 'adm_id', 'measurement', 'education_id', 'study_meas_id']
    return new_df

def add_measurement_id(df):
    new_df  = df.copy()
    new_df['measurement_id'] = np.arange(new_df.shape[0])
    ### adjusting the column order to get better visualization ###
    new_df = new_df[['measurement_id', 'study_meas_id', 'education_id','region_id', 'location_id', 'adm_id', 'year', 'measurement']]
    return new_df
```

```python
### function that compiles all the transformations ###
def transform_df(path, id, has_location, melt_columns):
    df = get_df(path)
    df = mapping_df(df, has_location)
    df = melt_df(df, melt_columns)
    df = get_education_id2(df)
    df['meas_group'] = df['variable'].apply(get_variable_group)
    df = get_study_id(df,id)
    df = get_studyMeas_id(df)
    df = remove_columns(df)
    df = rename_columns(df)
    df = add_measurement_id(df)
    return df
```

```python
def check_unique_values(df):
    print(f"study_meas_id uni: {df['study_meas_id'].unique()}")
    print(f"education_id uni: {df['education_id'].unique()}")
    print(f"region_id uni: {df['region_id'].unique()}")
    print(f"location_id uni: {df['location_id'].unique()}")
    print(f"adm_id uni: {df['adm_id'].unique()}")
    print(f"year uni: {df['year'].unique()}")
```

```python
def remove_null_rows(df):
    new_df = df.copy()
    new_df = new_df[(new_df['measurement'] != '--') & (new_df['measurement'] != '*')]
    try:
        new_df['measurement'] = new_df['measurement'].astype('float64')
    except ValueError as e:
        print(e)

    return new_df

```

### Concat all DataFrames ###

```python
### will store a unique dataframe with information of all years ###
def get_unique_df(list_dfs):
    unique_df = list_dfs[0]

    ### passing through dataframes and concatenating all of them ###
    for i in range(1, len(list_dfs)):
        unique_df = pd.concat([unique_df, list_dfs[i]], axis=0)

    return unique_df
```
